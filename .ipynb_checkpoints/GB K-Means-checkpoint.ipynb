```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs

n_samples = 1500
n_features = 2
random_state = 170
X, y = make_blobs(n_samples=n_samples, n_features=n_features, random_state=random_state)

features = ['x1', 'x2']
df = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=features)
df.head()
```

```python
sns.scatterplot(x='x1', y='x2', data=df)
```

# Model

```python
def create_centroid(n_k, n_features):
    centroids = []
    for i in range(n_k):
        c = []
        for j in range(n_features):
            c.append(np.random.uniform(0, 1))
        centroids.append(c)
    return centroids

def euclidean_distant(row_data, centroid):
    dist = 0
    for i in range(len(row_data)):
        dist += (row_data[i] - centroid[i])**2
        
    return np.sqrt(dist)

def get_cluster(data, centroids):
    cluster = []
    for i in range(len(data)):
        dists = []
        for j in range(len(centroids)):
            dist = euclidean_distant(data.iloc[i], centroids[j])
            dists.append(dist)
        
        min_value = min(dists)
        id_min = dists.index(min_value)
        cluster.append(id_min)

    return cluster

def new_centroids(df, n_k, cluster):
    data = df.copy()
    data['cluster'] = cluster
    data.groupby('cluster').mean()
    
    new_centroids = []
    for i in range(n_k):
        new_centroids.append(data.loc[i])
        
    return new_centroids

def hitung_sse(data, centroids):
    sse = 0
    for i in range(len(data)):
        for j in range(len(centroids)):
            dist = euclidean_distant(data.iloc[i], centroids[j])
            sse += dist**2

    return sse

def kmeans(n_k, n_feature, n_iter, df):
    data = df.copy()
    
    centroids = create_centroid(n_k, n_feature)
    
    for i in range(n_iter):
        cluster = get_cluster(data, centroids)
        
        centroids = new_centroids(data, n_k, cluster)
        
    sse = hitung_sse(data, centroids)
    return cluster, sse
```

# Main Program

```python
# main
labels, sse = kmeans(3, 2, 20, df)
```

```python
print('Sum of square error: ', sse)
```

```python
df_viz = df.copy()
df_viz['cluster'] = labels
sns.scatterplot(x='x1', y='x2', hue='cluster', data=df_viz)
```
